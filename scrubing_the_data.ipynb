```python
# Now let's get their id, by creatting a dictionry.
```

```python
sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(
    client_id=SPOTIPY_CLIENT_ID,
    client_secret=SPOTIPY_CLIENT_SECRET))
track_to_id = dict()
for track in unique_tracks:
    track_id = sp.search(q='artist:' + track[0] + ' track:' + track[1], type='track')
    if len(track_id['tracks']['items'])>0:
        track_to_id[track] = track_id['tracks']['items'][0]['id']
    else:
        track_to_id[track]= None
```

```python
# Also we will need the uri of the song
```

```python
sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(
    client_id=SPOTIPY_CLIENT_ID,
    client_secret=SPOTIPY_CLIENT_SECRET))
track_to_uri = dict()
for track in unique_tracks:
    track_uri = sp.search(q='artist:' + track[0] + ' track:' + track[1], type='track')
    if len(track_id['tracks']['items'])>0:
        track_to_uri[track] = track_id['tracks']['items'][0]['uri']
    else:
        track_to_uri[track]= None
```

```python
track_to_id
```

```python
# Now we can add then to our streaming history.
```

```python
for i in range(len(streaming_history)):
    streaming_history[i]['uri'] = track_to_uri[tuple([
        streaming_history[i]['artistName'], streaming_history[i]['trackName']
    ])]
    streaming_history[i]['id'] = track_to_id[tuple([
        streaming_history[i]['artistName'], streaming_history[i]['trackName']
    ])]
```

```python
# Now let's get the audo features of a song.
```

```python
sp.audio_features(streaming_history[0]['uri'])
```

```python
# Now lets get all of them for every song
for i in range(len(streaming_history)):
    if streaming_history[i]['uri'] is not None:
        audio_features = sp.audio_features(streaming_history[i]['uri'])
        streaming_history[i]['danceability'] = audio_features[0]['danceability']
        streaming_history[i]['energy'] = audio_features[0]['energy']
        streaming_history[i]['key'] = audio_features[0]['key']
        streaming_history[i]['loudness'] = audio_features[0]['loudness']
        streaming_history[i]['mode'] = audio_features[0]['mode']
        streaming_history[i]['speechiness'] = audio_features[0]['speechiness']
        streaming_history[i]['acousticness'] = audio_features[0]['acousticness']
        streaming_history[i]['instrumentalness'] = audio_features[0]['instrumentalness']
        streaming_history[i]['liveness'] = audio_features[0]['liveness']
        streaming_history[i]['valence'] = audio_features[0]['valence']
        streaming_history[i]['time_signature'] = audio_features[0]['time_signature']
        streaming_history[i]['duration_ms'] = audio_features[0]['duration_ms']
```

```python
# Now we must get the artist genre. For this I will firstly create a set of all artists.
```

```python
artists = set()
for track in streaming_history:
    artists.add(track['artistName'])
```

```python
artists
```

```python
artists = list(artists)
track_id = sp.search(q='artist:' + artists[0])
```

```python
track_id['tracks']['items']
sp.artist(track_id['tracks']['items'][0]['album']["artists"][0]["external_urls"]["spotify"])
```

```python
# Creatting the artist to genre dictionary
artist_to_genre = dict()
for artist in artists:
    track_id = sp.search(q='artist:' + artist)
    genres = track_id['tracks']['items']
    if len(genres) > 0:
        genres = genres[0]['album']["artists"]
        if len(genres) > 0:
            genres = sp.artist(genres[0]["external_urls"]["spotify"])['genres']
            artist_to_genre[artist] = genres
        else:
            artist_to_genre[artist] = None
    else:
        artist_to_genre[artist] = None
    artist_to_genre[artist] = genres
```

```python
artist_to_genre
```

```python
# Adding genres to every track
for i in range(len(streaming_history)):
    streaming_history[i]['genres'] = artist_to_genre[streaming_history[i]['artistName']]
```

```python
# Converting the genres list into different key-value pairs.
for i in range(len(streaming_history)):
    if streaming_history[i]['genres'] is not None:
        for genre in streaming_history[i]['genres']:
            streaming_history[i][genre] = 1
    del streaming_history[i]['genres']
```
